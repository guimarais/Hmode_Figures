```python
from ipfnpytools.plot import plot_signals
import ipfnpytools.aug_read as aug_read
#import matplotlib.pyplot as plt
#plt.style.use('/home/guimas/python/ipfnpytools/ipfnpytools/helvet2.mplstyle')
from getsig import getsig
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:950pt !important; }</style>"))
```

```python
shotnr = 34519
names = ["H-0", "H-5", "PECRH", "PNI", "PICRFc",r"D_tot",r"N_tot",r"Pre_OuDi",r"Pre_InDi","F12","F18"]
data = aug_read.many_signals(
    diagnostics=(2*["DCN"] + ["ECS","NIS","ICP"] + 2*["UVS"] + 2*["MSQ"] +2*["IOC"]),
    names=names,
    shots=[shotnr])
```

```python
fig, axes, fig_name = plot_signals(
    data=data, postfixes="overview", boxes=([0]*2 + [1]*3 + [2]*2 + [3]*2 + [4]*2),maintain_colors=True,
    legend_kw=dict(loc='best'),
    signal_multipliers=([1E-19] * 2 + [1E-6]*3 + [1E-22]*2 + [133.0]*2 + [1.0]*2),
    y_labels=[r"$\overline{n_e}$ [$10^{19}m^{-3}]$", r"[MW]", r"$[10^{22}e/s]$", r"[Pa]", r"[e/s]"], 
    y_mins=0.0,
    t_begin=1, t_end=8,
    title=r'#34532'    
)
```
